```python
npath = "fasta/Ecoli_neg.fa"
ppath = "fasta/Ecoli_pos.fa"
data = SequenceDinucProperties(npath, ppath)
```

```python
Xpos = data.get_positive_samples()
Xneg = data.get_negative_samples()
Y = data.getY()

print(Xpos.shape)
print(Xneg.shape)
print(Y.shape)
```

```python
def get_prop(df, i):
    return  pd.DataFrame(df[:,i,:,0], dtype='float')
```

```python
prop = 3
```

```python
plt.figure()
dfPos = get_prop(Xpos, prop)
ax = sns.violinplot(data=dfPos.iloc[:,41:61], palette="Set2", split=True, inner="quartile", scale="count", bw=.1)
plt.show()
```

```python
plt.figure()
dfNeg = get_prop(Xneg, prop)
ax = sns.violinplot(data=dfNeg.iloc[:,41:61], palette="Set2", split=True, inner="quartile", scale="count", bw=.1)
plt.show()
```
